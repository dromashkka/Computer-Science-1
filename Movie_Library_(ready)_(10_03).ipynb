In this project, a movie library system was created. There the information about movies, including their titles, release years, genres, ratings, and reviews,ews can be stored. The user can check the library, search movies by genres, look for details of movies by titles, add reviews, add reviews and clculate average rating

```python
#creating a list where movies will be stored

movie_library = []

# we have different movies: Harry Potter and the philosopher's stone, The Hobbit, Dune

# let's add them using command .append() as we have list. all information will be stored as dictionarry

movie_library.append({
# adding Harry Potter
    'Title': "Harry Potter and the philosopher's stone",
    'Year': 2001,
    'Genres': ['Fantasy', "Adventure", 'Family'],
    'Rating': (8.3, 325),
    'Reviews': {'Bob': 'my favorite movie'}
})

# adding The Hobbit
movie_library.append({
    'Title': "The Hobbit",
    'Year': 2012,
    'Genres': ['Fantasy', "Adventure", 'Action'],
    'Rating': (8.1, 627),
    'Reviews': {'Jake': 'it was good'}
})

# adding Dune
movie_library.append({
    'Title': "Dune",
    'Year': 2021,
    'Genres': ['Fantasy', "Adventure", 'Action', 'Drama'],
    'Rating': (7.7, 384),
    'Reviews': {'Alex': 'interesting and fascinating'}
})
```

```python
# writing menu
while True:
#while true will allow printing the menu after every point
#while loop runs while a condition is True. we did not give a condition; therefore, it is always True, and only break can stop the loop which is in 6. Leave
  print("What would you like to do next?")
  print("1. Check existing Library")
  print("2. Search Movies by genre")
  print("3. Look for details by Title")
  print("4. Add a Review to a Movie")
  print("5. Calculate avarage rating")
  print("6. Leave")

# asking for number from menu. what we are going to do
#in variable a, store it as an integer to operate after and use for 'if'
  a = int(input())

## Menu 1: Create code to display all movies inside the library
  #writing the list of movies, because a == 1
  if a == 1:
    print("---------Your Library---------")
    print('')
    #creating an empty variable so that after we could make it a str for genre and make it possible to write beautifully in one str
    g = ''
    for movie in movie_library: # with for we are going through each movie and "unpack" the dictionary with it, so we will see a beautiful menu
      print("Title:", movie['Title']) #with key 'movie' we get access to Title and get it to print
      print("Year:", movie['Year']) #same with Year
      for i in range(len(movie['Genres'])): #as have many Genras will use for to look and unpack each of them, len calculates number of objects in list genres
        g = g + str(movie['Genres'][i]) + ' ' #to write genres in one str
      print("Genres:", g)
      g = '' #deleting all information from g, so we can use it again
      print("Rating:", movie['Rating'][0],'('+str(movie['Rating'][1]), 'reviews)') #printing rating, movies is a key to Rating, but [0] and [1] give us positions of overall rating and reviews
      print("")
      print("-----------------------------")
      print('')


## Menu 2: Search Movies by Genre
  #looking at genre as a == 2
  if a == 2:
    attempts = 0
    #has only 3 attempts to write correctly, attempts - counter of attempts
    while attempts < 3: #use while to ask for input and check if a number of attempts is less than 3
      print("Which genre are you interested in? Start with capital letter")
      genre = input()
      c = 0 #need for count of movies with this genre

      # this is for beautiful menu
      print('')
      print("---------Found movies---------")
      print('')

    #checking genre
      for movie in movie_library: #looking at each movie in list using for.
        for k in range(len(movie['Genres'])): #looking at the section genres in movie using for and the number of genras
          if str(movie['Genres'][k]) == str(genre): #checking if suits
            c +=1
            #writing information about movie
            print("Title:", movie['Title'])
            print("Year:", movie['Year'])
            for i in range(len(movie['Genres'])):
              g = g + str(movie['Genres'][i]) + ' '
            print("Genres:", g)
            g = ''
            print("")
            print("-----------------------------")
            print('')

      if c == 0: #if there is no that genre
        attempts += 1
        print("Something is wring")
        print("Attempts left:", 3 - attempts)
      else:
        break # breaks if c != 0, because no need for more attempts and movies are found

## Menu 3: View Details of a Specific Movie
  #searching by title for details
  if a == 3:
    title_movie = input('title: ') #asking for title to search

    print('')
    print("---------Found movies---------")
    print('')

    #using for loop --if to look at each movie in list and look at its name
    for movie in movie_library:
      #if the information suits, then print all info about movie
      if title_movie in movie['Title']:
          print("Title:", movie['Title'])
          print("Year:", movie['Year'])
          for k in range(len(movie['Genres'])):
            g = g + str(movie['Genres'][k]) + ' '
          print("Genres:", g)
          g = ''
          print("")
          print("-----------------------------")
          print('')
          break
      else:
        print('Not in library')
        print("-----------------------------")
        print('')

## Menu 4: Add a Review to a Movie
#add review
  if a == 4:
    rev_title = input('Which movie would you like to add a review to? ' ) #asking for the title of movie
    reviewer = input("What is the name of the reviewer? ") #asking for reviewer's name
    rating = int(input('What is your rating out of 10? '))
    review = input('What is your review? ') #asking for the review itslf

    #finding movie in the library using for to look at all titles
    for movie in movie_library:
      if movie['Title'] == rev_title:  # checking if title in the library and if the input match
        movie['Reviews'][reviewer] = review #getting access to reviews in movie, then add reviewer and add review

        #new rating
        old_rating, old_num_reviews = movie['Rating']  # Tuple Unpacking the current rating and number of reviews
        new_num_reviews = old_num_reviews + 1 #+1 because we add 1 more review - our review
        new_rating = ((old_rating * old_num_reviews) + rating) / new_num_reviews  # New rating with a review score of 10
        movie['Rating'] = (new_rating, new_num_reviews)  # Update the movie with the new rating and number of reviews

        #print the movie name and reviewer
        print('')
        print("-----------------------------")
        print(f"Review added to {rev_title} by {reviewer} successfully")
        print("-----------------------------")
        print('')
        break  #stop because movie is found

## Menu 5: Calculate and Display the Average Rating
#calculate avarage rating as a == 5
  if a == 5:
    #creating variables to use
    new_total_rating = 0
    new_total_reviews = 0
    #looking at all movies to add them to new total rating and reviews
    for movie in movie_library:
      rating, reviews = movie['Rating'] #Tuple Unpacking the current rating and number of reviews
      new_total_rating += rating * reviews
      new_total_reviews += reviews

    if new_total_reviews > 0:
      print('')
      print("-----------------------------")
      print(new_total_rating / new_total_reviews)
      print("-----------------------------")
      print('')
    else:
      print('No reviews')

## Menu 6: stop the program
#if a == 6 and user wants to leave
  else:
    break #to stop the program
```
