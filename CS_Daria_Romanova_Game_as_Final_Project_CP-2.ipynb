1. Planning the game

Rules:
- The Tic-Tac-Toe game is intended for two players; therefore, it is made for
humans, not computers and players.

- Each player puts X or O. Three in a column, in line, or in a diagonal - WIN

Game logic:
- It is going to have a board 3*3 with numbers from 1 to 9 in each cell. These numbers are going to represent the position taken.

- After the players input X or O, numbers are replaced with X and O

- After 4 turns, the game starts checking for a winner, using stored combinations for winning

- Game is going to have a foolproof: it is going to print a message if the input is invalid: letters and other signs, cell is taken, numbers are not between 1 and 9

My steps:

- Create a board, store there information about game (as a list), draw it, using for loop

- Take input from the players. Check if it is valid, using functions, while and if loops

- Create a function that will be win/draw checker, using tuples, if and while loops

- Create a function-mechanism that is going to rule the board, whose turn is it, if there is win or draw. Use for it

```python
#print the title of the game
print("*" * 10, " Tic Tac Toe ", "*" * 10)

#create the game board as a list of numbers from 1 to 9 - positions that players will choose

board = []

for i in range(1, 10):
  board.append(i)

#fnction to draw the board in a grid format

def draw_board(board):
  print("-" * 13) #top
  for i in range(3): #range 3 as we have 3 lines
      print("|", board[0 + i * 3], "|", board[1 + i * 3], "|", board[2 + i * 3], "|")
      print("-" * 13)

# Function to take the player's input and place their mark. it checks if the input was valid: not letter, not taken place, not other numbers
def take_input(player_token): #player's token is 'X' or 'O'
  #while true will allow printing the menu after every point
  #while loop runs while a condition is True. we did not give a condition; therefore, it is always True, and only break can stop the loop.

  while True:
    print(" ")
    player_answer = input("Where would you like to put " + player_token + "? ")
    print(" ")
    #checking for numbers, not other signs, using .isdigit()

    if player_answer.isdigit():
        player_answer = int(player_answer)
        #check if number is between 1 and 9
        if 1 <= player_answer <= 9:
          #checking if place is taken. here we use board[player_answer-1] because our bord starts from 1 and not from 0. board[player_answer] would give us next number, not the one we want
          if str(board[player_answer - 1]) not in 'XO':
            board[player_answer - 1] = player_token #change cell to player's answer
            break  #exit the loop after a successful move
          else:
            print(" ")
            print("That cell is already taken!")
            print(" ")
        else:
          print(" ")
          print("Invalid number. Choose a number between 1 and 9.")
          print(" ")
    else:
       print(" ")
       print("Invalid input. Please enter a number.")
       print(" ")

#function to check for a win/draw and print message
def check_win_draw(board):
    #use tuple to store there all possible winning combinations
    win_coord = (
        (0, 1, 2), (3, 4, 5), (6, 7, 8),      # line
        (0, 3, 6), (1, 4, 7), (2, 5, 8),      # column
        (0, 4, 8), (2, 4, 6)                  # diagonal
    )

    #check each combination using for loop
    for combination in win_coord:
      #comparing digits
      #looking for combination first using board[combination], then get access to signs with combination[0/1/2]
        if board[combination[0]] == board[combination[1]] == board[combination[2]]:
            return board[combination[0]]  #return the winner because digits are the same

#function for turns. it will show who will be next, who will win, if there is draw and the end of the game.
def turn(board):

    counter = 0  #counter of turns

    #using while loop, check the state of the board
    #while true will allow printing the menu after every point
    #while loop runs while a condition is True. we did not give a condition; therefore, it is always True, and only break can stop the loop.
    while True:
        draw_board(board)  #keep showing the board
        #using if to find out whose turn it is
        if counter % 2 == 0: #X have odd numbers of turns and O have even numbers
            take_input("X")  #player X turn
        else:
            take_input("O")  #player O turn
        counter += 1 #as turn was done

        #using if to check the winner
        if counter > 4: #put more than 3 X or O for one player
            who_win = check_win_draw(board) #checking if someone won. in that def we returned sign: X or O. if no wins, nothing is returned
            if who_win == 'X' or who_win == 'O':
                print(" ")
                print(who_win, "won the game!")
                print(" ")
                break
        if counter == 9:
            print(" ")
            print("Draw!")  #all cells are filled
            print(" ")
            break
    draw_board(board)  # Final board state

#start the game
turn(board)

#final message
print(" ")
print('Game Over!')
print(" ")
```
